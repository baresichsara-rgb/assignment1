## 1. Understanding & Breakdown



Input: an integer inputted by the user, programmed as the variable “n”.
Output: the program must return a 1-indexed string array. Therefore, the list of strings must begin at 1, and until the integer “n”.

Conditions:
If the integer input is divisible by 3 and 5, the program must return “FizzBuzz”.
If the integer is divisible only by 3, the program must return “Fizz”.
If the integer is divisible by 5, the output is “Buzz”.
Where the number is not divisible in the list of strings, by any of the above conditions, the output is “i” or in other words, the string itself.

Constraints: the constraints listed here are: 1 <= n <= 104
This implies that the “n” integer must always be a minimum of one, and it cannot be more than 10 000.

	Steps to Solve the Fizz Buzz Problem:
1. Insert the string array to hold the list. “n” is now inputted as an integer.  
2. Create a for loop, in which the variable “i” will be declared and inputted as 1, so that the loop will account for the numbers beginning from 1, up until “n” is when the for loop will stop.
2. Inside of the loop are “if” statements… First, it is calculated if “i” (“i” and not n, since “i” is every passing number being checked up until “n”) can be divisible by 3 and 5. If this statement is fulfilled, “FizzBuzz” will be added for place of the integer.
3. Then, there are “else if” statements. If “i” is divisible by 3, then “Fizz” is added in place of the integer. The next “else” repeats similarly, if “i” is divisible by 5, rather, the string “Buzz” will be added instead
4. Then, the last “else” adds the remaining numbers themselves as a string… If they are not divisible by 3 or 5, the actual numbers are added.
5. Lastly, the for loop is closed, and the list of string arrays is returned as a result.

```python
# Draft code.
class Solution {
public:
    vector<string> fizzBuzz(int n) {
        vector<string> list;

        for (int i = 1; i <= n; i++){
            if (i % 3 == 0 && i % 5 == 0) {
                list.add("FizzBuzz");
            } else if (i % 3 == 0){
                list.add("Fizz");
            } else if (i % 5 == 0){
                list.add("Buzz");
            } else {
                list.add(to_string(i));
            }
        }
        return list;

    }
};
```

```python
# Final working solution
# [Disclaimer: I used an additional solution to review and understand why my code was not running.]
class Solution {
public:
    vector<string> fizzBuzz(int n) {
        vector<string> list;

        for (int i = 1; i <= n; i++){
            if (i % 15 == 0) {
                list.push_back("FizzBuzz");
            } else if (i % 3 == 0){
                list.push_back("Fizz");
            } else if (i % 5 == 0){
                list.push_back("Buzz");
            } else {
                list.push_back(to_string(i));
            }
        }
        return list;

    }
};
```
