-10^9 <= target <= 10^9

Steps to Solve the Two Sum Problem:

1. Declare a hash table, begin to build the hash table by using a for loop.
2. The for loop will loop through the array to identify each of them numbers and add them to the hash table.
3. Create amother for loop that will identify what the numbers are in the array that add up to the target, and the output for those numbers. This is completed by subracting each of the numbers, one at a time, from the target, which equals the other number (the complement).
4. If the complement is existing in the hash table, a count will be added for that particular number for the index. If the complement in not existing, the return is empty, and the code continues to loop until it finds a complement.
5. Return the correct indices.

## 3. Final Correct Code




```python
# Final Working Solution
# [AI Disclaimer. I used ChatGPT to define the errors: "How am I able to improve this code? Are there errors?"]
class Solution {
public:
    vector<int> twoSum(vector<int>& nums, int target) {
        unordered_map<int, int> numMap;

        for (int i = 0; i < nums.size(); ++i){
            int complement = target - nums[i];

            if (numMap.find(complement) != numMap.end()){
                return {numMap[complement], i};
            }

            numMap[nums[i]] = i;
        }
        return {};


    }
};
```
