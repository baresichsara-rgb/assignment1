## 1. Understanding & Breakdown



Input: an integer inputted by the user, programmed as the variable â€œnâ€.
Output: the program must return a 1-indexed string array. Therefore, the list of strings must begin at 1, and until the integer â€œnâ€.

Conditions:
If the integer input is divisible by 3 and 5, the program must return â€œFizzBuzzâ€.
If the integer is divisible only by 3, the program must return â€œFizzâ€.
If the integer is divisible by 5, the output is â€œBuzzâ€.
Where the number is not divisible in the list of strings, by any of the above conditions, the output is â€œiâ€ or in other words, the string itself.

Constraints: the constraints listed here are: 1 <= n <= 104
This implies that the â€œnâ€ integer must always be a minimum of one, and it cannot be more than 10 000.

	Steps to Solve the Fizz Buzz Problem:
1. Insert the string array to hold the list. â€œnâ€ is now inputted as an integer.  
2. Create a for loop, in which the variable â€œiâ€ will be declared and inputted as 1, so that the loop will account for the numbers beginning from 1, up until â€œnâ€ is when the for loop will stop.
2. Inside of the loop are â€œifâ€ statementsâ€¦ First, it is calculated if â€œiâ€ (â€œiâ€ and not n, since â€œiâ€ is every passing number being checked up until â€œnâ€) can be divisible by 3 and 5. If this statement is fulfilled, â€œFizzBuzzâ€ will be added for place of the integer.
3. Then, there are â€œelse ifâ€ statements. If â€œiâ€ is divisible by 3, then â€œFizzâ€ is added in place of the integer. The next â€œelseâ€ repeats similarly, if â€œiâ€ is divisible by 5, rather, the string â€œBuzzâ€ will be added instead
4. Then, the last â€œelseâ€ adds the remaining numbers themselves as a stringâ€¦ If they are not divisible by 3 or 5, the actual numbers are added.
5. Lastly, the for loop is closed, and the list of string arrays is returned as a result.

```python
# Draft code.
class Solution {
public:
    vector<string> fizzBuzz(int n) {
        vector<string> list;

        for (int i = 1; i <= n; i++){
            if (i % 3 == 0 && i % 5 == 0) {
                list.add("FizzBuzz");
            } else if (i % 3 == 0){
                list.add("Fizz");
            } else if (i % 5 == 0){
                list.add("Buzz");
            } else {
                list.add(to_string(i));
            }
        }
        return list;

    }
};
```

```python
# Final working solution
# [Disclaimer: I used an additional solution to review and understand why my code was not running.]
class Solution {
public:
    vector<string> fizzBuzz(int n) {
        vector<string> list;

        for (int i = 1; i <= n; i++){
            if (i % 15 == 0) {
                list.push_back("FizzBuzz");
            } else if (i % 3 == 0){
                list.push_back("Fizz");
            } else if (i % 5 == 0){
                list.push_back("Buzz");
            } else {
                list.push_back(to_string(i));
            }
        }
        return list;

    }
};
```
